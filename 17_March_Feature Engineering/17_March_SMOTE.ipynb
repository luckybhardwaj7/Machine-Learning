# SMOTE Technique

SMOTE is an oversampling technique where the synthetic samples are generated for the minority class. This algorithm helps to overcome the overfitting problem posed by random

```python
# Creating a New Imbalanced Dataset
```

```python
from sklearn.datasets import make_classification
```

```python
# X is Independent Feature as it is Input
# Y is Dependent feature as it is Output
x,y=make_classification(n_samples=1000,
                    n_features=2,
                    n_redundant=0,
                    n_clusters_per_class=1,
                    weights=[0.90],
                    random_state=1)
```

```python
x,y
```

```python
import pandas as pd
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```

```python
final_df['target'].value_counts()
# Will generate Truelyimbalanced Data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
#Transform the Dataset
oversample=SMOTE()
x,y=oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
```
